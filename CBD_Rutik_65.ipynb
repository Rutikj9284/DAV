Plotly :- Plotly is a Python graphing library used for creating interactive, publication-quality graphs and charts. It provides a high-level API that allows users to create a wide range of plots easily. plotly.express is a high-level interface for Plotly, a Python graphing library that makes interactive, publication-quality graphs online.

# Import Libraries

```python
import pandas as pd
import plotly.express as px
```

# Import Dataset

```python
data = pd.read_csv("stock_data.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
```

# Plotting Different Charts

1.Line chart

```python
fig1 = px.line(data, x='Date', y='Close', title='Trend of Close Prices Over Time')
fig1.show()
```

2.Bar Chart

```python
data['Year'] = data['Date'].dt.year
avg_volume = data.groupby('Year')['Volume'].mean().reset_index()
fig2 = px.bar(avg_volume, x='Year', y='Volume', title='Average Volume per Year')
fig2.show()
```

Box Plot

```python
fig3 = px.box(data, y='High', title='Distribution of High Prices')
fig3.show()
```

Scatterplot

```python
fig4 = px.scatter(data, x='High', y='Low', title='Relationship between High and Low Prices')
fig4.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Candlestick(
    x=data['Date'],
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    close=data['Close']
)])

# Update the layout
fig.update_layout(
    title='Candlestick Chart of Stock Prices',
    xaxis_title='Date',
    yaxis_title='Price',
    xaxis_rangeslider_visible=False
)

# Show the plot
fig.show()
```
